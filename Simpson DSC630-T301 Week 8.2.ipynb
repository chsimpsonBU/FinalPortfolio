Plot the data with proper labeling and make some observations on the graph.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("us_retail_sales.csv")


# Get the sum of sales for each month
monthly_sales = data.sum()

# Create a bar graph of the monthly sales
plt.bar(monthly_sales.index, monthly_sales)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales (billions of dollars)')
plt.title('US Retail Sales')

# Show the plot
plt.show()
```

I decided to transform the data from columns to DATE rows with Sales before modeling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# use melt to unpivot the data
melted_df = pd.melt(data, id_vars=['YEAR'], value_vars=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'], var_name='MONTH', value_name='SALES')

# concatenate the year and month columns into a single column
melted_df['DATE'] = pd.to_datetime(melted_df['YEAR'].astype(str) + '-' + melted_df['MONTH'], format='%Y-%b')

# drop the YEAR and MONTH columns
final_df = melted_df.drop(['YEAR', 'MONTH'], axis=1)

# set the DATE column as the index
final_df = final_df.set_index('DATE')

final_df

```

Dropping the NaN values that will be predicted later. 

```python
final_df = final_df.dropna()
```

Using a Linear Regression model

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Convert index to datetime type
final_df.index = pd.to_datetime(final_df.index)

train = final_df.loc[final_df['DATE'].dt.year < 2020]
test_start_date = '06-01-2020'
test_end_date = '07-01-2021'
test = final_df.loc[(final_df['DATE'] >= test_start_date) & (final_df['DATE'] < test_end_date)]


# Build a predictive model using the training set
X_train = train.index.month.values.reshape(-1, 1)
y_train = train['SALES']
model = LinearRegression()
model.fit(X_train, y_train)

# Use the model to predict monthly retail sales on the test set
X_test = test.index.month.values.reshape(-1, 1)
y_test = test['SALES']
y_pred = model.predict(X_test)

# Report the RMSE of the model predictions on the test set
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```
