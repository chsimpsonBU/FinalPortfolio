```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset from Seaborn library
iris = sns.load_dataset('iris')

# Creating a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heat Map - Iris Dataset Python')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

#Beyonce's next three tour stops on the Renaissance Tour
data = pd.DataFrame({
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Lat': [40.7128, 34.0522, 41.8781],
    'Lon': [-74.0060, -118.2437, -87.6298],
    'Population': [8600000, 4000000, 2700000]
})

# Creating the spatial chart
fig = px.scatter_mapbox(data, lat='Lat', lon='Lon', size='Population', color='Population', size_max=30,
                        mapbox_style='carto-positron', zoom=3, hover_name='City')
fig.update_layout(title='Spatial Chart - City Populations Python', mapbox=dict(center=dict(lat=37.0902, lon=-95.7129)))
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Weather in Omaha
data = pd.DataFrame({
    'Categories': ['Monday', 'Tuesday', 'Wednesday', 'Thursday'],
    'Values': [91, 96, 95, 93]
})

# Create the lollipop chart
plt.figure(figsize=(8, 6))
plt.stem(data['Categories'], data['Values'], basefmt=' ', use_line_collection=True)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Lollipop Chart - Python')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
