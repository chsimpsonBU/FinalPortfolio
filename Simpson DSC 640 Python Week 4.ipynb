```python
import csv

# Open the CSV file
with open('us-postage.csv', 'r') as file:
    # Create a CSV reader object
    reader = csv.reader(file)
    
    # Skip the header row if it exists
    next(reader)
    
    # Create empty lists to store the data
    years = []
    prices = []
    
    # Read each row and extract the year and price
    for row in reader:
        year = int(row[0])
        price = float(row[1])
        years.append(year)
        prices.append(price)
```

```python
import matplotlib.pyplot as plt
import squarify

# Tree Map
plt.figure(figsize=(8, 6))
plt.title('Python- Treemap')
squarify.plot(sizes=prices, label=years, alpha=0.8)
plt.axis('off')
plt.show()

# Area Chart
plt.figure(figsize=(8, 6))
plt.title('Python- Area Chart')
plt.fill_between(years, prices, color='skyblue', alpha=0.6)
plt.plot(years, prices, color='royalblue')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()

# Stacked Area Chart
plt.figure(figsize=(8, 6))
plt.title('Python- Stacked Area Chart')
plt.stackplot(years, prices, labels=years, colors='skyblue', alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.show()

# Step Chart
plt.figure(figsize=(8, 6))
plt.title('Python- Step Chart')
plt.step(years, prices, where='mid')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```
