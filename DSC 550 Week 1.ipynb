### Load the dataset as a Pandas data frame

```python
import pandas as pd

# Load the CSV file into a pandas dataframe
GameDataRaw = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')

# Print the dataframe to verify that it loaded correctly
GameDataRaw.head(10)
```

### Find the dimensions (number of rows and columns) in the data frame.

```python
num_rows, num_cols = GameDataRaw.shape
print("Number of Rows: ", num_rows)
print("Number of Columns: ", num_cols)
```

### What do these two numbers represent in the context of the data?
The number of rows represents the observed games, and the columns represent attributes about the game.

### Find the top five games by critic score.

```python
top_five = GameDataRaw.sort_values('Critic_Score', ascending=False).head(5)

# I like to sort by row appearance here.
top_fiveSorted = top_five.sort_index()

# Print the top five games
print(top_fiveSorted['Name'])
```

### Find the number of video games in the data frame in each genre.

```python
counts_by_genre = GameDataRaw.groupby('Genre').size()

# Let's do ascending order
counts_by_genre = counts_by_genre.sort_values()

# Print the counts by genre
print(counts_by_genre)
```

### Find the first five games in the data frame on the SNES platform.

```python
snes_games = GameDataRaw[GameDataRaw['Platform'] == 'SNES']

# Select the first five games in the filtered dataframe
first_five = snes_games.head(5)

# Let's order them
first_fiveSorted = first_five.sort_index()


# Print the first five games
print(first_five['Name'])
```

### Find the five publishers with the highest total global sales.


```python
# Global sales is already on the sheet. Validated a few rows and the data is correct. 

top_five_global_sales = GameDataRaw.sort_values('Global_Sales', ascending=False).head(5)

# Print the top five games
print(top_five_global_sales['Name'])
```

### Create a new column in the data frame that calculates the percentage of global sales from North America. Display the first five rows of the new data frame.

```python
na_sales_percent = GameDataRaw['NA_Sales'] / GameDataRaw['Global_Sales'] * 100

# Add the new column to the dataframe
GameDataRaw['na_sales_percent'] = na_sales_percent

# Display the first five rows of the new dataframe
GameDataRaw.head(5)
```

### Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
import numpy as np

# Replace non-numerical strings with NaNs in the user score column
GameDataRaw['User_Score'] = pd.to_numeric(GameDataRaw['User_Score'], errors='coerce')

# Calculate the median user score
median_user_score = GameDataRaw['User_Score'].median()

# Replace all NaN entries in the user score column with the median value
GameDataRaw['User_Score'].fillna(median_user_score, inplace=True)


GameDataRaw.head(5)
```
